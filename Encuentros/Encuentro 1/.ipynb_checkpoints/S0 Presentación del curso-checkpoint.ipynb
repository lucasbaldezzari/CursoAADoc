### Encuentro 1

1. Presentacin del curso. Instalacin y preparacin del entorno de trabajo. 
2. Conceptos fundamentales del lenguaje Python. Tipos y estructuras de datos bsicas. Estructuras de control. Funciones y clases. Manejo de archivos. Importacin de libreras. Documentacin y sitios de consulta. 
3. Computacin numrica. Matrices y arreglos n-dimensionales. ndices y seleccin. Operaciones principales. Importacin y exportacin de datos. Vectorizacin y broadcasting. 
4. Graficacin. Tipos de grficas y sus elementos. Exportacin.

### Encuentro 2

1. Introduccin a *machine learning* (aprendizaje automtico). Motivacin del aprendizaje automtico.
2. Tipos de problemas en machine learning: clasificacin, regresin y clustering. Algoritmos y funcionamiento. 
3. Figuras de mrito y evaluacin del desempeo. Esquema de particionado de datos.

### Encuentro 3

1. Exploracin de datos y preprocesamiento. Manejo de datos estructurados. ndices y seleccin. Exploracin. 
2. Procesamiento y operaciones con seales. Transformada de Fourier y filtrado.
3. Manipulacin de imgenes. Mtodos bsicos de procesamiento. Transformaciones de caractersticas y vectores objetivo. Aumentacin. Desbalance.

### Encuentro 4

1. Caso de estudio I: Aplicacin de *machine learning* en un problema no supervisado (agrupamiento de datos) y un problema de regresin. 
2. Se abordar la resolucin completa de un problema, con el bloque de procesamiento en cascada, desde la lectura del corpus de seales, preprocesado de las mismas, definicin del modelo apropiado, entrenamiento, evaluacin del desempeo, exploracin y seleccin de hiper-parmetros, validacin del modelo y anlisis de los resultados.

### Encuentro 5

1. Caso de estudio II: Aplicacin de *machine learning* en un problema de clasificacin. 
2. Se abordar la resolucin completa de un problema, con el bloque de procesamiento en cascada, desde la lectura del corpus de seales, preprocesado de las mismas, definicin del modelo apropiado, entrenamiento, evaluacin del desempeo, exploracin y seleccin de hiper-parmetros, validacin del modelo y anlisis de los resultados.

### Encuentro 6

1. Introduccin a redes neuronales. Algoritmo de retropropagacin.

### Encuentro 7

1. Caso de estudio III: Aplicacin de *deep learning* a la clasificacin de seales temporales.

### Encuentro 8
