## Objetivo general

Adquirir intuiciones básicas para aplicar técnicas de *machine learning* a señales, imágenes y datos estructurados, utilizando el lenguaje de programación Python y su ecosistema de desarrollo.

## Objetivos particulares

Se espera que el alumno pueda:
1. Adquirir conocimientos básicos de *machine learning* para abordar problemas de agrupación de datos, regresión y clasificación.
2. Comprender la importancia y adquirir las destrezas para realizar el preprocesamiento y adecuación de los datos al modelo elegido.
3. Implementar, entrenar y validar algoritmos fundamentales de *machine learning*.
4. Implementar, entrenar y validar redes neuronales básicas usando técnicas basadas en *deep learning*.

### Encuentro 1

1. Presentación del curso. Instalación y preparación del entorno de trabajo. 
2. Conceptos fundamentales del lenguaje Python. Tipos y estructuras de datos básicas. Estructuras de control. Funciones y clases. Manejo de archivos. Importación de librerías. Documentación y sitios de consulta. 
3. Computación numérica. Matrices y arreglos n-dimensionales. Índices y selección. Operaciones principales. Importación y exportación de datos. Vectorización y broadcasting. 
4. Graficación. Tipos de gráficas y sus elementos. Exportación.

### Encuentro 2

1. Introducción a *machine learning* (aprendizaje automático). Motivación del aprendizaje automático.
2. Tipos de problemas en machine learning: clasificación, regresión y clustering. Algoritmos y funcionamiento. 
3. Figuras de mérito y evaluación del desempeño. Esquema de particionado de datos.

### Encuentro 3

1. Exploración de datos y preprocesamiento. Manejo de datos estructurados. Índices y selección. Exploración. 
2. Procesamiento y operaciones con señales. Transformada de Fourier y filtrado.
3. Manipulación de imágenes. Métodos básicos de procesamiento. Transformaciones de características y vectores objetivo. Aumentación. Desbalance.

### Encuentro 4

1. Caso de estudio I: Aplicación de *machine learning* en un problema no supervisado (agrupamiento de datos) y un problema de regresión. 
2. Se abordará la resolución completa de un problema, con el bloque de procesamiento en cascada, desde la lectura del corpus de señales, preprocesado de las mismas, definición del modelo apropiado, entrenamiento, evaluación del desempeño, exploración y selección de hiper-parámetros, validación del modelo y análisis de los resultados.

### Encuentro 5

1. Caso de estudio II: Aplicación de *machine learning* en un problema de clasificación. 
2. Se abordará la resolución completa de un problema, con el bloque de procesamiento en cascada, desde la lectura del corpus de señales, preprocesado de las mismas, definición del modelo apropiado, entrenamiento, evaluación del desempeño, exploración y selección de hiper-parámetros, validación del modelo y análisis de los resultados.

### Encuentro 6

1. Introducción a redes neuronales. Algoritmo de retropropagación. 
2. *Deep learning* (aprendizaje profundo). Operaciones con tensores. Programación diferenciable. Arquitecturas neuronales básicas.
3. Preparación de datos. Modos de entrenamiento. Validación de modelos. Figuras de mérito. Optimización de parámetros.

### Encuentro 7

1. Caso de estudio III: Aplicación de *deep learning* a la clasificación de señales temporales.
2. Se abordará la resolución completa de un problema, desde la lectura del corpus de señales, preprocesado de las mismas, definición del modelo apropiado, entrenamiento, evaluación del desempeño, revisiones del modelo y análisis de los resultados.

### Encuentro 8

1. Caso de estudio IV: Aplicación de *deep learning* a la clasificación de imágenes. 
2. Se abordará la resolución completa de un problema, desde la lectura del corpus de imágenes, preprocesado del mismo, definición del modelo, entrenamiento, evaluación del desempeño, revisiones del modelo y análisis de los resultados.
