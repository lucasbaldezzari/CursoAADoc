## Graficación

La herramienta es el intermediario entre el usuario y la gráfica:

  - **Frontend**: interacción con el usuario, permite **definir** una gráfica
  - **Backend**: interacción con el sistema, permite **representar** una gráfica

## Figura simple

```python
import numpy as np
x = np.linspace(-np.pi, np.pi, 256) 
y_cos, y_sin = np.cos(x), np.sin(x)

plt.plot(x, y_sin);
plt.plot(x, y_cos);
```

### Personalizar la figura

```python
plt.plot(x, y_cos, color="blue", linewidth=5, linestyle="--");   # Estilo, grosor
plt.plot(x, y_sin, color="red",  linewidth=2.5, linestyle="-");  # Color
plt.xlim(x.min() * 1.5, x.max() * 1.5);                          # Limites
plt.ylim(y_cos.min() * 1.25, y_cos.max() * 1.25);
```

### Guardar la gráfica en un archivo

```python
plt.plot(x, y_cos, x, y_sin);
plt.legend(["cos(x)" , "sen(x)"], fontsize="xx-large", loc="upper left");
plt.title("Seno y coseno", size="xx-large")
plt.text(0.2, 0.5, r'$\sin(x)$', fontsize="x-large", horizontalalignment='center')
plt.savefig("figura_simple.png")  # svg, png, ...
```

## Múltiples figuras

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4)) # 2 columnas
axes[0].plot(x, y_cos)  # prestar atencion al indexado de los axes
axes[1].plot(x, y_sin);
```

## Múltiples figuras

```python
# axes es una matriz de 2 x 2
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 4))  
axes[0, 0].plot(x, y_cos)  # prestar atencion al indexado de los axes
axes[0, 1].plot(x, y_sin)
axes[1, 0].plot(x, -y_cos)
axes[1, 1].plot(x, -y_sin);
```

## Más allá de matplotlib

Hay varias librerías especializadas en gráficas interactivas, para sitios web, etc.
Por ejemplo `bokeh`

```python
# ! pip install bokeh
```

```python
from bokeh.io import output_notebook, show, output_file
from bokeh.plotting import figure

output_notebook()  # Permite ver la salida en la notebook
```

```python
p = figure(plot_width=400, plot_height=300, title="My Line Plot")

p.line(x, y_sin, legend_label="sen(x)", color="blue", line_width=2)
p.line(x, y_cos, legend_label="cos(x)", color="green", line_width=2)
p.legend.click_policy="hide" 
p.legend.location = "top_left"

output_file("ejemplo_bokeh.html")  # Figura en html standalone
show(p) # show the results
```
