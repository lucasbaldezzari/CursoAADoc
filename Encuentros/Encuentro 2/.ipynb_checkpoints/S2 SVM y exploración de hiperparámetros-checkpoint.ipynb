## Exploración de hiperparámetros para SVM

Para ver detalles de implementación y ejemplos, consultar documentación en https://scikit-learn.org/stable/modules/svm.html#svm

## Cargamos un dataset de prueba

En este caso usaremos el mismo de iris, pero filtramos solo las clases 1:versicolor y 2:virgínica, que son más dificiles de separar

```python
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt 
import numpy as np 

X, y = load_iris(return_X_y=True)


# Filtramos las etiquetas que no son 0 (incluye 1 y 2).
# Además usaremos sólo las caraceterísticas 0 y 2
ind = np.where(y != 0)[0] 
X = X[ind,:][:, (0, 2)]
y = y[ind]

plt.figure(figsize=(6,3))
for l in np.unique(y):
    ind = y == l
    plt.plot(X[ind, 0], X[ind, 1], "o", label=f"clase{l}")

plt.xlabel("Característica 0 (longitud de sépalo)")
plt.ylabel("Característica 2 (longitud del pétalo)")
plt.legend();
    
```

## Graficando la frontera de decisión
Una vez entrenado el modelo, a cada punto del plano podemos asignarle un puntaje, algo similar a una probabilidad de que el punto sea de una clase u otra. Esto nos sirve para visualizar como será la frontera de decisión para el problema. 

Notar que no dividimos en train/test porque solo estamos estudiando como es la frontera de decisión

```python
def plot_decision_function(classifier, axis, title, X, y):
    """Grafica la distribución de puntajes por clase y vectores de soporte
    
    Parameters
    ----------
    classifier : SVC
        Clasificador entrenado.
    axis : Axis
        Ejes a graficar.
    title : str
        Título del gráfico
    X : numpy.ndarray
        Características
    y : numpy.ndarray
        Etiquetas
    """
    
    x1 = np.min(X[:, 0])
    x2 = np.max(X[:, 0])
    y1 = np.min(X[:, 1])
    y2 = np.max(X[:, 1])
    xx, yy = np.meshgrid(np.linspace(x1, x2, 200), np.linspace(y1, y2, 200))

    
    Z = classifier.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    axis.contourf(xx, yy, Z, alpha=0.75, levels=4)
    axis.scatter(X[:, 0], X[:, 1], c=y, alpha=0.9,
                 cmap=plt.cm.bone, edgecolors='black')

    # Vectores de soporte
    axis.scatter(classifier.support_vectors_[:, 0], classifier.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
    
    axis.axis('off')
    axis.set_title(title)
```

## Entrenamiento y exploración de C
Usamos la clase SVC (SVM para clasificación). Entrenamos tres modelos, cada uno con un valor diferentes de C. 

Notar como cambia la cantidad de vectores de soporte (puntos resaltados) y la frontera decisión.


```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Creamos la figura y grilla de graficacion (la misma en todos los casos)
fig, ax = plt.subplots(1,3, figsize=(15,5))
    
    
for k, c in enumerate([1e-3, 1, 1e10]):
    model = SVC(C=c, kernel="linear")
    model.fit(X, y)
    acc = accuracy_score(y, model.predict(X))
    plot_decision_function(model, ax[k], f"C={c}, Acc={acc: .3f}", X, y)    
```

## Entrenamiento de SVM con kernel RBF y exploración de C y gamma

Repetimos el análisis anterior, esta vez cambiando el kernel de la SVM  y cambiando sus diferentes hiperparámetros

```python
fig, ax = plt.subplots(3, 5, figsize=(16,10))
    
for i, g in enumerate([1e-5, 1e-3, "scale", 1, 100]):
    for j, c in enumerate([1e-3, 1, 1e10]):
        model = SVC(C=c, gamma=g, kernel="rbf")
        model.fit(X, y)
        acc = accuracy_score(y, model.predict(X))
        plot_decision_function(model, ax[j, i], f"C={c: .0e}, g={g}, Acc={int(100*acc)}", X, y)
```

## Selección de hiperparámetros.

Las pruebas anteriores son útiles a fines de entender el algoritmo y la influencia de sus hiperparámetros. En una aplicación real, hacemos una exploración de los hiperparámetros en términos de una medida de desempeño, para asi elegir con que configuración trabajaremos. 

Notar que es fundamental **usar particiones de entrenamiento y prueba**, de otra forma estaremos midiendo solo que tan bueno es el modelo para memorizar datos

```python
from sklearn.model_selection import train_test_split

# hiperparámtros a evaluar
gamma_list = 10**np.arange(-10, 3, dtype=float)
C_list = 10**np.arange(-2, 5, dtype=float)

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=.2)   

results = np.zeros((len(gamma_list), len(C_list)))
# guardamos los resutlados en entrenamiento solo para compararlos con los de 
# prueba
results_train = np.zeros_like(results)

for i, g in enumerate(gamma_list):
    for j, c in enumerate(C_list):
        model = SVC(C=c, gamma=g, kernel="rbf")
        # Entrenamiento
        model.fit(Xtrain, ytrain)
        # Evaluación en entrenamiento (solo para comparar)
        results_train[i, j] = accuracy_score(ytrain, model.predict(Xtrain))
        # Evaluación 
        results[i, j] = accuracy_score(ytest, model.predict(Xtest))
```

```python
plt.figure(figsize=(15,10))
plt.imshow(results)
plt.xlabel("C")
plt.xticks(np.arange(len(C_list)), C_list)
plt.ylabel("Gamma")
plt.yticks(np.arange(len(gamma_list)), gamma_list)
plt.colorbar();

for i in range(results.shape[0]):
    for j in range(results.shape[1]):
        plt.text(j, i, results[i, j], va='center', ha='center')
```

¿Y que pasa si hacemos el mismo análisis con los resultados en la partición de entrenamiento?

```python
plt.figure(figsize=(15,10))
plt.imshow(results_train)
plt.xlabel("C")
plt.xticks(np.arange(len(C_list)), C_list)
plt.ylabel("Gamma")
plt.yticks(np.arange(len(gamma_list)), gamma_list)
plt.colorbar();

for i in range(results_train.shape[0]):
    for j in range(results_train.shape[1]):
        plt.text(j, i, results_train[i, j], va='center', ha='center')
```

## Comentario final

Además de usar particiones de entrenamiento y prueba, aquí es donde se hace evidente la importancia de tener al menos tres particiones: *entrenamiento*, *optimización* y *prueba*.

- *Entrenamiento*: Se usa para entrenar cada modelo y combinación de hiperparámetros.
- *Optimización*:  Se usa para evaluar el funcionamiento de cada modelo, con fines de elegir la mejor combinación de hiperparámetros.
- *Prueba*: Nos dará una estimación insesgada del desempeño real del modelo elegido. La usamos solo para evaluar el mejor modelo 

En este caso, la partición de *prueba* es necesaria porque, si se exploraron muchas combinaciones de hiperparámetros, es posible que al elegir el mejor modelo hayamos **sobreajustado** el modelo a los datos de *optimización*. 

Los resultados en entrenamiento y optimización solo sirven a fines del desarrollo. Por ejemplo, para saber si el modelo esta funcionando bien, o que tan estables son los resultados para diferentes hiperparámetros. El resultado que debemos informar es siempre el de *prueba*.
